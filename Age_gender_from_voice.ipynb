```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # visualizing data
import seaborn as sns # visualizing data with stunning default theme
import sklearn # contain algorithms
import warnings
warnings.filterwarnings('ignore')

# load dataset from input directory
df = pd.read_csv("/content/cv-valid-train.csv")
df[df['age'].notna()].head()
```

```python
#we extract the columns that we think useful are
df = df[['filename','age','gender']]
#To clean the data we remove the sample with NaN attribute values.
data = df[df['age'].notna() & df['gender'].notna()]
data.reset_index(inplace=True, drop=True)
data.head()
```

```python
#data['gender'] = pd.to_numeric(data['gender'],errors = 'coerce')
#data.dtypes
#if the below code fail to covert gender datatype and values then uncomment above code
cleanup_nums = {"gender": {"male":1,"female":0,"other":0.5}}
data = data.replace(cleanup_nums)
data.head()
```

```python
data[data['gender']==0.5].head()
```

```python
data.drop(data[data['gender']==0.5].index,inplace=True)
```

```python
data.head(228)
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.head(228)
```

```python
data.info()
```

```python
age_encode = {
  "twenties": 2,
  "thirties": 3,
  "fourties": 4,
  "fifties": 5,
  "sixties": 6,
  "seventies": 7,
  "eighties": 8,
  "teens": 1
  }
```

```python
data['age'] = data['age'].map(age_encode)
```

```python
data.head()
```

```python
import os
import librosa
import numpy as np

def extract_features(file_path):
    file_path = os.path.join("/content/cv-valid-train", file_path)
    y, sr = librosa.load(file_path, sr=16000)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfccs_scaled = np.mean(mfccs.T, axis=0)
    return mfccs_scaled

# Apply feature extraction to all files
data['features'] = data['filename'].apply(extract_features)
```

```python
data.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
```

```python
# Prepare the data
X = np.array(data['features'].tolist())
y_gender = np.array(data['gender'].tolist())
y_age = np.array(data['age'].tolist())
```

```python
y_age
```

```python

# Build the model
input_layer = Input(shape=(13,))
x = Dense(128, activation='relu')(input_layer)
x = Dropout(0.3)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.3)(x)

gender_output = Dense(1, activation='sigmoid', name='gender_output')(x)
age_output = Dense(1, name='age_output')(x)

model = Model(inputs=input_layer, outputs=[gender_output, age_output])
model.compile(optimizer='adam', loss={'gender_output': 'binary_crossentropy', 'age_output': 'mse'}, metrics={'gender_output': 'accuracy', 'age_output': 'mae'})

model.summary()
```

```python
# Train the model
model.fit(X, {'gender_output': y_gender, 'age_output': y_age}, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model
loss, gender_loss, age_loss, gender_accuracy, age_mae = model.evaluate(X, {'gender_output': y_gender, 'age_output': y_age})
print(f"Gender Accuracy: {gender_accuracy}, Age MAE: {age_mae}")
```

```python
# Save the model
model.save('age_gender_detection_model.keras')
```

```python
# Load the saved model
loaded_model = tf.keras.models.load_model('age_gender_detection_model.keras')

def predict_age_gender(file_path):
    features = extract_features(file_path)
    features = np.expand_dims(features, axis=0)
    features = np.expand_dims(features, axis=-1)
    gender_pred, age_pred = loaded_model.predict(features)

    gender = "Female" if gender_pred[0][0] < 0.5 else "Male"
    #age = age_pred[0][0]
    if age_pred[0][0] < 2:
      age = "Teens"
    elif age_pred[0][0] < 3:
      age = "Twenties"
    elif age_pred[0][0] < 4:
      age = "Thirties"
    elif age_pred[0][0] < 5:
      age = "Fourties"
    elif age_pred[0][0] < 6:
      age = "Fifties"
    elif age_pred[0][0] < 7:
      age = "Sixties"
    elif age_pred[0][0] < 8:
      age = "Seventies"
    else:
      age = "Eighties"

    return gender, age

# Example usage
file_path = '/content/cv-other-test/cv-other-test/sample-000014.mp3'
gender, age = predict_age_gender(file_path)
print(f"Predicted Gender: {gender}, Predicted Age: {age}")
```
